## üßΩ Data Cleaning Order (Recommended Workflow)

When cleaning the Netflix dataset, it's ideal to follow a logical sequence to improve data quality efficiently. Below is the recommended order:

1. **Quality ‚û§ Completeness**

   - Cross-verify values with trusted external sources where possible  
   - Correct known inaccurate values (e.g., wrong durations)

5. **Quality ‚û§ Consistency**  
   - Standardize formats (e.g., consistent use of "min", "Season(s)", proper spacing in lists)  
   - Ensure categories or labels are uniform across the dataset

<b><font color="white"><h3>üîç Function to Handle Missing Values in Specific Columns:</b></font>

```python
def Unknown_Name(df, col):
    """
    Replaces NaN values in the specified column with 'Unknown_<column_name>'.

    Parameters:
    df (DataFrame): The input DataFrame.
    col (str): The column name where NaNs need to be replaced.

    Returns:
    DataFrame: The updated DataFrame with NaNs in the column replaced by 'Unknown_<col>'.
    """
    unknown_value = f"Unknown_{col}"
    df[col] = df[col].fillna(unknown_value)
    return df
```

```python
for column in ['director', 'cast', 'country']:
    df = Unknown_Name(df, column)
```

üîπ **Handling Missing Values in 'director', 'cast', and 'country' Columns:**

The provided function `Unknown_Name` is used to replace `NaN` values in specific columns with a placeholder string. In this case, the function is applied to the **'director'**, **'cast'**, and **'country'** columns.

üìù **Insight:**
- The missing values in the **'director'**, **'cast'**, and **'country'** columns are now replaced with placeholder values like `'Unknown_director'`, `'Unknown_cast'`, and `'Unknown_country'`.
- This helps in preserving rows while making it clear which entries are missing data, avoiding the need to drop rows and thus keeping the dataset intact.
- It is particularly useful for categorical data where missing entries might represent unknown information rather than irrelevant data.
- Now, any missing director, cast, or country information is clearly labeled, making future analysis smoother and more transparent.

This approach ensures no data is lost, and the missing values are easily identifiable.

### <b><font color="white"><h3>üîç Unstacking and Exploding Columns to Handle Multiple Values:</b></font>

```python
def unstack_col(df, col):
    # Ensure values are strings or keep NaNs as-is
    df[col] = df[col].apply(lambda x: x.strip() if isinstance(x, str) else x)

    # Split only strings by comma
    df[col] = df[col].apply(lambda x: x.split(',') if isinstance(x, str) else x)

    # Explode and remove NaNs from that column
    df = df.explode(col).dropna(subset=[col]).reset_index(drop=True)

    return df
```

```python
# Apply to all required columns
for column in ['director', 'cast', 'country', 'listed_in']:
    df = unstack_col(df, column)
```

üìù Insight:

The unstack_col function is designed to handle columns that contain multiple values in a single cell, such as lists or comma-separated values. Here's what happens:

1. String Stripping: It strips any leading or trailing spaces from string values.
2. Splitting Values: For each string, it splits the values by a comma, creating a list of values.
3. Exploding: The explode() method then transforms each value in the list into a separate row, effectively "unstacking" the values.
4. Removing NaNs: After exploding, any resulting NaN values are dropped to ensure the cleaned dataset only contains valid entries.

By applying this function to columns like 'director', 'cast', 'country', and 'listed_in', we can ensure that each individual value is represented in a separate row, facilitating more accurate analysis and improving the overall data quality.
